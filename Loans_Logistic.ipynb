```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
loans = pd.read_csv('loans.csv')
```

```python
loans.head()
```

```python
loans = loans.drop(['id'], axis = 1)
```

```python
loans.describe(include='all')
```

```python
loans.isnull().sum()
```

```python
loans_dropna = loans.dropna(subset = ['emplength','annualinc','delinq2yrs','inqlast6mths','openacc','revolutil','totalacc'])
```

```python
loans_dropna.isnull().sum()
```

```python
loans_dropna.describe(include='all')
```

```python
loans_dropna.columns
```

```python
loans_dropna.dtypes
```

```python
loans_dropna = loans_dropna.astype({"creditpolicy":'object'})
```

```python
loans_dropna_status_bar = sns.barplot(x="targetloanstatus", y="loanamnt", data=loans_dropna, 
                               estimator=lambda x:len(x)/len(loans_dropna)*100)
loans_dropna_status_bar.set(xlabel = "Target Loan Status", ylabel="Percentage")
totals = []
for i in loans_dropna_status_bar.patches:
    totals.append(i.get_height())
total = sum(totals)
for i in loans_dropna_status_bar.patches:
    loans_dropna_status_bar.text(i.get_x()+.3, i.get_height()-5, 
                          str(round((i.get_height()/total)*100,2))+'%',
                          fontsize=10, color='white')
```

```python
loans_final = pd.get_dummies(loans_dropna, prefix = ['creditpolicy', 'term', 'grade', 'emplength', 
                   'homeownership', 'purpose', 'verificationstatus'], 
                         columns = ['creditpolicy', 'term', 'grade', 'emplength', 
                   'homeownership', 'purpose', 'verificationstatus'],
                            drop_first = 'TRUE')
```

```python
loans_final.columns
```

```python
loans_final.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(loans_final.drop(['targetloanstatus'], axis = 1), 
                                                    loans_final['targetloanstatus'], test_size=0.30, random_state=0)
```

```python
len(x_train)/len(loans_final)
```

```python
len(x_test)/len(loans_final)
```

```python
len(y_train)/len(loans_final)
```

```python
len(y_test)/len(loans_final)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["Variables"] = x_train.columns
vif["VIF Factor"] = [variance_inflation_factor(x_train.values, j) for j in range(x_train.shape[1])]
vif.round(4)
```

```python
x_train_drop_coll = x_train.drop(['loanamnt','intrate'], axis = 1)
vif = pd.DataFrame()
vif["Variables"] = x_train_drop_coll.columns
vif["VIF Factor"] = [variance_inflation_factor(x_train_drop_coll.values, j) for j in range(x_train_drop_coll.shape[1])]
vif.round(4)
```

```python
import statsmodels.api as sm
model1 = sm.Logit(y_train,x_train)
result1 = model1.fit()
result1.summary2()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression(solver='lbfgs')
logreg.fit(x_train, y_train)
y_predit = logreg.predict(x_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(x_test, y_test)))
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_predit)
print(confusion_matrix)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predit))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt 
logit_roc_auc = roc_auc_score(y_test, logreg.predict(x_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(x_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
